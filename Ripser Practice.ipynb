```python
import numpy as np
from ripser import ripser
from persim import plot_diagrams
from sklearn import datasets
import matplotlib.pyplot as plt
import persim
np.random.seed(4)
alldata = np.random.random((1000,2))
plt.plot(alldata[:,0],alldata[:,1],'o')
```

```python
diagrams = ripser(alldata)['dgms']
#print(ripser(alldata))
plot_diagrams(diagrams, show=True)
```

```python
np.random.seed(4)
alldata2 = np.random.random((1000,2))

data=np.asarray([x for x in alldata2 if not(0.2<min(x[0],x[1]) and max(x[0],x[1])<0.8)])
plt.plot(data[:,0],data[:,1],'o')
```

```python
diagrams2 = ripser(data)['dgms']
plot_diagrams(diagrams2, show=True)
```

```python
data3 = 0.5+ (datasets.make_circles(n_samples=1000)[0] + 5 * datasets.make_circles(n_samples=1000)[0])/12
plt.plot(data3[:,0],data3[:,1],'o')
```

```python
diagrams3 = ripser(data3)['dgms']
plot_diagrams(diagrams3, show=True)
```
